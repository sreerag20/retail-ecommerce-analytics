```python
# Stage 7 — Power BI Export Preparation
## Export all views and segments as CSV files for Power BI dashboard
```

```python
from sqlalchemy import create_engine
import pandas as pd
import os

engine = create_engine(
    'mysql+mysqlconnector://portfolio_user:YourPass123@localhost/retail_portfolio'
)

exports_path = "D:/Portfolio/Projects/retail_portfolio/data/exports"
os.makedirs(exports_path, exist_ok=True)

# All exports for Power BI
exports = {
    'fact_orders'        : "select * from fact_orders",
    'dim_customers'      : "select * from dim_customers",
    'dim_products'       : "select * from dim_products",
    'dim_locations'      : "select * from dim_locations",
    'dim_sales_reps'     : "select * from dim_sales_reps",
    'vw_orders_enriched' : "select * from vw_orders_enriched",
    'vw_monthly_revenue' : "select * from vw_monthly_revenue",
    'vw_rfm'             : "select * from vw_rfm",
    'vw_cohort'          : "select * from vw_cohort",
}

for filename, query in exports.items():
    df = pd.read_sql(query, engine)
    fpath = f"{exports_path}/{filename}.csv"
    df.to_csv(fpath, index=False)
    print(f"✅ Exported {filename}.csv - {len(df):,} rows, {len(df.columns)} columns")
```
